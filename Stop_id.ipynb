```python
import numpy as np
import pandas as pd
```

```python
stop = pd.read_excel('file2.xlsx', sheet_name = 'files.csv')
trips = pd.read_excel('file2.xlsx', sheet_name = 'trips.csv')
routes = pd.read_excel('file2.xlsx', sheet_name = 'routes.csv')
```

```python
stop_id = []
trip_id = []
route_id = []
def plswork():
  for i in range(0,len(routes['route_id'])):
    for j in range(0,len(trips['route_id'])):
      if(routes['route_id'][i]==trips['route_id'][j]):
        if(trips['trip_id'][j].find('Up')!=-1):
          for k in range(0,len(stop['trip_id'])):
            if(stop['trip_id'][k]==trips['trip_id'][j]):
              stop_id.append(stop['stop_id'][k])
              trip_id.append(trips['trip_id'][j])
              route_id.append(routes['route_id'][i])
        elif(trips['trip_id'][j].find('Down')!=-1):
          for k in range(0,len(stop['trip_id'])):
            if(stop['trip_id'][k]==trips['trip_id'][j]):
              stop_id.append(stop['stop_id'][k])
              trip_id.append(trips['trip_id'][j])
              route_id.append(routes['route_id'][i])  
                
plswork()
```
