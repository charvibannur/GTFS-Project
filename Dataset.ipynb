```python
import numpy as np
import pandas as pd
dataset=pd.read_csv('/content/finalSlot2.csv')
df  = dataset.drop_duplicates()
df['No of trips']=0
```

```python

for j in range(0,len(df['stop_id_modified_from'])):
  count=0
  for i in range(0,len(df['stop_id_modified_from'])):
    if((df['stop_id_modified_from'][i]==df['stop_id_modified_from'][j]) & (df['stop_id_modified_to'][i]==df['stop_id_modified_to'][j])):
      count+=1
  df['No of trips'][j]=count
df.to_csv('slot_7_00-7_20_trips.csv')
```
